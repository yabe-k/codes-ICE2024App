```python
import numpy as np
import pandas as pd

import os
```

```python
filenames = os.listdir()

country_files = [filename for filename in filenames if filename.startswith("country_matrix_") and filename.endswith("npy")]
country_files.sort()
country_files

field_files = [filename for filename in filenames if filename.startswith("field_matrix_") and filename.endswith("npy")]
field_files.sort()
field_files
```

```python
country_data = []
field_data = []

for country_file, field_file in zip(country_files, field_files):
  country_data.append(np.load(country_file))
  field_data.append(np.load(field_file))

country_data = np.concatenate(country_data, axis=0)
field_data = np.concatenate(field_data, axis=0)

print(country_data.shape)
print(field_data.shape)
```

```python
# 行・列名を読み込む
country_row_col_names = pd.read_csv("country_row_col_names.csv", index_col=0)
field_row_col_names = pd.read_csv("field_row_col_names.csv", index_col=0)
```

```python
country_data_mean = country_data.mean(axis=0)
country_data_std = country_data.std(axis=0)

field_data_mean = field_data.mean(axis=0)
field_data_std = field_data.std(axis=0)

country_data_mean = pd.DataFrame(country_data_mean, index=country_row_col_names.index, columns=country_row_col_names.columns)
country_data_std = pd.DataFrame(country_data_std, index=country_row_col_names.index, columns=country_row_col_names.columns)

field_data_mean = pd.DataFrame(field_data_mean, index=field_row_col_names.index, columns=field_row_col_names.columns)
field_data_std = pd.DataFrame(field_data_std, index=field_row_col_names.index, columns=field_row_col_names.columns)
```

```python
field_data_mean.to_csv("sample_mean_field.csv")
field_data_std.to_csv("sample_std_field.csv")
country_data_mean.to_csv("sample_mean_country.csv")
country_data_std.to_csv("sample_std_country.csv")

from google.colab import files
files.download("sample_mean_field.csv")
files.download("sample_std_field.csv")
files.download("sample_mean_country.csv")
files.download("sample_std_country.csv")
```
